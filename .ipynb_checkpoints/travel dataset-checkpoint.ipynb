```python
# import libaries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# read csv file 
df=pd.read_csv("Top Indian Places to Visit.csv")
print(df.shape)#print rows and columns of dataframe.
df.info()#print dataframe  columns ,no-null values ,count and datatype of dataset.
```

# Missing Data
Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.


```python
# looking for missing values in dataset.
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{}- {}%'.format(col,round(pct_missing*100)))
    print(df.shape)
```

```python
# subsetting
df_travel = df[[
#     'Unnamed: 0', 
    'Zone', 'State', 'City', 'Name', 'Type',
    'Establishment Year', 'time needed to visit in hrs',
    'Google review rating', 'Entrance Fee in INR',
    'Airport with 50km Radius', 'Weekly Off', 'Significance',
    'DSLR Allowed', 'Number of google review in lakhs',
    'Best Time to visit'
]].copy()
```

```python
# categorical features
list_categorical_features = [
    'Zone', 'State', 'City', 'Name', 'Type', 
    'Establishment Year', 'Airport with 50km Radius', 'Weekly Off', 
    'Significance', 'DSLR Allowed', 'Best Time to visit'
]
df_travel[list_categorical_features] = df_travel[list_categorical_features].astype('category')
df_travel.info()
```

```python
#missing values
df_travel.isnull().sum()
```

```python
# showing missing value through heat map.
plt.figure(figsize=(10,8))

cols = df_travel.columns[:30]
colours = ['#000099', '#ffff00'] # specify colours: yellow - missing. blue - not missing
sns.heatmap(df_travel[cols].isna(), cmap=sns.color_palette(colours))
print(df.shape)
```

# unecessary data
I have already removed one of the column 'unnamed' from the dataset.iw ant keep reamining columns for my analysis.

```python
### Inconsistent Data
# simplifying by combining similar catergories.
df_travel['Type']=df_travel['Type'].replace({'Temples': 'Temple', 'Tombs': 'Tomb'})
# Weekly Off Correction: Assuming 'Yes' is an error and without clear indication of what it should be,
# we can set it to NaN
df_travel['Weekly Off']=df_travel['Weekly Off'].replace('Yes',None)
# 3. Best Time to Visit Consistency: Normalize values
df_travel['Best Time to visit'] = df_travel['Best Time to visit'].str.strip()
df_travel['Best Time to visit'] = df_travel['Best Time to visit'].replace({'Anytime': 'All'})
```

```python
# save the cleaned dataset into csv
df_travel.to_csv("travel_data_cleaned.csv",index=False)
```

```python
# Check to see that it workded
df_cleaned = pd.read_csv("travel_data_cleaned.csv")

# Want to see if the dataframe was properly imported and see what kind of data is stored in each column.
df_cleaned.info()
df_cleaned.head()
```
